```python
import youtube_dlc
import pandas as pd
import numpy as np
```

```python
queries = ["datascience", "data+science","ciencia+de+dados", "machine+learning", "aprendizado+de+maquina", "kaggle"]
```

```python
ydl = youtube_dlc.YoutubeDL({"ignoreerrors": True})
```

```python
"ytsearchdate10:{}".format('ciencia+de+dados')
```

```python
%%time
resultados = []
for query in queries:
    r = ydl.extract_info("ytsearchdate4000:{}".format(query), download=False)
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    resultados += r['entries']
```

```python
resultados = [e for e in resultados if e is not None]
len(resultados)
```

```python
df = pd.DataFrame(resultados)

df['upload_date'] = pd.to_datetime(df["upload_date"])
df['tempo_desde_pub'] = (pd.to_datetime("2020-11-17") - df['upload_date']) / np.timedelta64(1, 'D') 
```

```python
df_filtrado = df[['id','webpage_url','title','playlist','view_count','like_count','dislike_count','average_rating','subscriber_count','description','categories','tags','upload_date','tempo_desde_pub', 'channel_url', 'thumbnail','width','height','resolution','fps']]
df_filtrado ['id'] = 'watch?v='+df_filtrado ['id']
df_filtrado .rename(columns={'id':'link','playlist':'query'}, inplace=True)
```

```python
df_filtrado .to_csv('dados/raw_data_sem_labels.csv', index=None)
```
